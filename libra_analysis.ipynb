```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
libra = pd.read_csv("Libra Database_ 2021-05-20.csv", delimiter=";")

print(libra.head())
print(libra.tail())
print(libra.dtypes)
```

```python
#libra.date = pd.to_datetime(libra.date, format="%Y-%m-%d")
libra.date = pd.to_datetime(libra.date, format="%Y-%m-%d %H:%M")

print(libra.head())
print(libra.tail())
print(libra.dtypes)
```

```python
sns.boxplot(y=libra.weight)
plt.show()
```

```python
plt.scatter(x=libra.date, y=libra.weight)
#plt.figure(figsize=(3, 3))
plt.rcParams['figure.dpi'] = 100
plt.show()
```

```python
plt.clf()
#plt.figure(figsize=(3, 2))
fig, ax = plt.subplots()
fig.set_figwidth(10)

ax.plot(libra.date, libra.weight, color="red")
ax.set_xlabel("Date")
ax.set_ylabel("Weight")

ax2 = ax.twinx()
ax2.plot(libra.date, libra['body fat'])
ax2.set_ylabel("Body Fat")
plt.show()
```

```python
libra['year'] = pd.DatetimeIndex(libra.date).year
print(libra.head())

years = libra.year.unique()
#print(years)

fig, ax = plt.subplots(len(years))

for i, year in enumerate(years):
    #ax[i].scatter(x=libra.date[libra.year==year], y=libra.weight[libra.year==year])
    ax[i].plot(libra.date[libra.year==year], libra.weight[libra.year==year])
    #start_date = str(year)+'01-01'
    #end_date = str(year)+'12-31'
    
    ax[i].set_xlim(pd.Timestamp(str(year)+'-01-01'), pd.Timestamp(str(year)+'-12-31'))
    ax[i].set_ylim(165, 185)

fig.set_size_inches(18.5, 25)
fig.savefig('test2png.png', dpi=100)
plt.show()
    
```
